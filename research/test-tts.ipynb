```python
import TTS
from TTS.utils.synthesizer import Synthesizer

        # self,
        # tts_checkpoint: str = "",
        # tts_config_path: str = "",
        # tts_speakers_file: str = "",
        # tts_languages_file: str = "",
        # vocoder_checkpoint: str = "",
        # vocoder_config: str = "",
        # encoder_checkpoint: str = "",
        # encoder_config: str = "",
        # vc_checkpoint: str = "",
        # vc_config: str = "",
        # model_dir: str = "",
        # voice_dir: str = None,
        # use_cuda: bool = False,

# Paths to configuration and checkpoint files
tts_config_path = "TTS/tts_models/en/tacotron2-DDC/config.json"
tts_checkpoint = "TTS/tts_models/en/tacotron2-DDC/checkpoint.pth"

# Initialize the TTS synthesizer
synthesizer = Synthesizer(
    tts_config_path = "TTS/tts_models/en/tacotron2-DDC/config.json",
    tts_checkpoint = "TTS/tts_models/en/tacotron2-DDC/checkpoint.pth"
    )

# Define the text to be synthesized
text = "Hello, welcome to Mozilla TTS!"

# Generate speech
waveform, sample_rate = synthesizer.tts(text)
```

```python
# Save the generated speech to a file
import numpy as np
import soundfile as sf

sf.write('output.wav', waveform.numpy(), sample_rate)
print("Speech saved to output.wav")
```
